```python
cfg = {
    "Dataset_Address_Evaluate": "/content/UCMerced_LandUse",
    "Trained_Model_Path": "/content/drive/MyDrive/MRSAC INTERNSHIP FILES /NasNet_Mobile_New_Loss3.117-0.9952(UCMerced_80_20).h5",  # pre-trained model
    "Eval_Batch_Size": 1  # Batch size
}
from google.colab import files
import cv2
import numpy as np
from keras.models import load_model

# Load the uploaded image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)
image = cv2.resize(image, (256, 256))  # Resize the image to match the input size of the model
image = image.astype(np.float32) / 255.0  # Normalize the pixe6l values

NASNET_Model = load_model(cfg["Trained_Model_Path"], custom_objects={"New_loss3": "categorical_crossentropy"})

# Make prediction
prediction = NASNET_Model.predict(np.expand_dims(image, axis=0))

# Define class labels
class_labels = [
    'agricultural', 'airplane', 'baseballdiamond', 'beach', 'buildings',
    'chaparral', 'denseresidential', 'forest', 'freeway', 'golfcourse',
    'harbor', 'intersection', 'mediumresidential', 'mobilehomepark',
    'overpass', 'parkinglot', 'river', 'runway', 'sparseresidential', 'storagetanks'
]

# Print the predicted class
predicted_class_index = np.argmax(prediction)
print("Predicted class index:", predicted_class_index)

# Get the predicted class
predicted_class_label = class_labels[predicted_class_index]
print("Predicted class label:", predicted_class_label)
```
